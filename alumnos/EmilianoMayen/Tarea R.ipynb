## Crecimiento poblacional de 1950-2050

### La gráfica muestra cómo la cifra de la población total ha ido, y espera siga, incrementando a lo largo del periodo en consideración (1970-2050)

```r
p <-estatal %>%
group_by(AÑO) %>%
summarise(POB_TOTAL = sum(POBLACION), M = sum(POBLACION[SEXO == 'Mujeres']), H = sum(POBLACION[SEXO == 'Hombres']))
ggplot(p, aes(AÑO, POB_TOTAL)) +
geom_point(size = 1, colour = 'cadetblue3') +
labs(title = 'Crecimiento poblacional de 1950-2050', y = 'Población', x = 'Año')
```

Observamos que si se toma la poblacion total acumulada (sumando todos los estados) a través de los años, tiene pendiente positiva pero disminuyendo 'Primera mitad de gráfica normal'

## Distribución de la población por edad y sexo. Pirámide poblacional

La siguiente gráfica es una pirámide poblacional, conformada por gráficas de barras dispuestas horizontalmente, cuya longitud es proporcional a la cantidad de personas que representa la edad y sexo de la población.

```r
estatal %>% 
group_by(SEXO, EDAD)%>%
summarise(POB_TOTAL = sum(POBLACION)) %>%
ggplot(aes(x= EDAD, y = ifelse(SEXO == 'Mujeres', -POB_TOTAL, POB_TOTAL), fill = SEXO)) +
geom_bar(stat = 'identity') +
scale_x_continuous(breaks = seq(from = 0, to= 110)) +
coord_flip() + theme_bw() + labs(title = 'Distribución de la población por edad y sexo', y = 'Población por sexo', x = 'Edad')
```

```r
datap <- estatal %>% 
group_by(AÑO, SEXO, EDAD)%>%
summarise(POB_TOTAL = sum(POBLACION)) %>%
filter(AÑO == 2000)
hombres <- datap %>% 
filter(SEXO == 'Hombres') %>% 
mutate(POB_TOTAL = POB_TOTAL * -1, SEXO = 'Hombres')
mujeres <- datap %>% 
filter(SEXO == 'Mujeres') %>% 
mutate(SEXO = 'Mujeres')
piramide <- rbind(hombres, mujeres) %>%
mutate(POBLACION = abs(POB_TOTAL))

text = ~paste("Población: ", POB_TOTAL)

plot_ly(piramide, x= ~POB_TOTAL, y = ~EDAD, color = ~SEXO) %>%
add_bars(orientation = "h",
        hoverinfo = "y+text+name", text = ~paste('Población:', POBLACION),
        colors = c("yellowgreen", "hotpink2")) %>%
  layout(bargap = 0.1, barmode = "overlay", 
         title = "Pirámide poblacional 2000", 
         xaxis = list(tickmode = "array", 
                      tickvals = c(-1200000, -1000000, -500000, 0, 500000, 1000000, 1200000),
                      ticktext = c("1200k", "1000k", "500k", "0", "500k", "1000k", "1200k"), 
                      title = "Población"), 
         yaxis = list(title = "Edad"))
```
