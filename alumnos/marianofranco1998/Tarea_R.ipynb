**Parte 1:**

De la base de datos de proyecciones poblacionales de CONAPO (aquí la base), hacer al menos 5 visualizaciones para entender el crecimiento poblacional. Se tienen muchas variables y categorías por explorar, se espera un análisis y contexto de las gráficas así como una interpretación.

```r
CONAPO = read.csv("pob_mit_proyecciones.csv", fileEncoding = "Latin1")
df <- data.frame(CONAPO)
```

**Visualización #1**

```r
pregunta1_1 <- filter(df, ENTIDAD == "República Mexicana") %>%
  group_by(AÑO, SEXO) %>%
  summarise(POBLACION_TOTAL = sum(POBLACION, na.rm = TRUE))
```

```r
ggplot(pregunta1_1, aes(AÑO,POBLACION_TOTAL,fill=SEXO))+
geom_point(size=0.1)+
geom_line(size = 1, color = "blue")+
facet_wrap(~SEXO,ncol=1)+
ggtitle("Crecimiento de Población por Sexo")+
theme(legend.position="none")+
xlab("Año")+
ylab("Población Total")
```

**Visualización #2**

```r
pregunta1_2 <- filter(df, ENTIDAD == "República Mexicana", EDAD == 0) %>%
  group_by(AÑO) %>%
  summarise(TASA_DE_NATALIDAD = sum(POBLACION, na.rm = TRUE))

pregunta1_2
```
