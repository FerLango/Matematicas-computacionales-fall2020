La graficación es una manera eficiente de resumir y mostrar infromación. Cuatro frases que resumen el verdadero propósito de la visualización de datos:
 - Las gráficas son para análisis cualitativos o descriptivos y quizá semi cuantitaticos, nunca para análisis profundo cuantitativo. 
 - Las gráficas son para comparaciones y/o comparaciones entre grupos, no para acceder a cantidades particulares.
 - Las gráficas son para impactar, de primera instancia, mover percepción, transmitir información. No se debería pensar mucho para llegar a la conclusión. 
 - Las gráficas deberían de reportar análisis de datos trabajado, fino y cuidadoso. No se debería intentar que la gráfica sea en sí el análisis. Las gráficas están para fortalecer el análisis, no para fundamentario.
